```python
import pandas as pd
import numpy as np
import tensorflow.keras as tf
```

```python
df=pd.read_csv("BreastCancerGrouped.csv")
df.value_counts().reset_index().to_csv('df.csv')
df.value_counts()
```

```python
df
```

```python
Y= df['class']
X= df.iloc[:,1:2]
Y
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.30,random_state=0)
```

```python
Y_train
```

```python
X_train
```

```python
X_test
```

```python
Y_test
```

```python
model = tf.models.Sequential()
model.add(tf.layers.Dense(16,input_dim=1,activation='relu'))  
model.add(tf.layers.Dense(32,activation='relu')) 
model.add(tf.layers.Dense(1,activation='sigmoid'))  
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=["accuracy"])
```

```python
model.fit(X_train,Y_train,epochs=500)
```

```python
model.summary()
```

```python
Y_pread=model.predict(X)
Y_pread
```

```python
Y_pread=Y_pread.round()
Y_pread
```

```python
model.evaluate(X_test,Y_test)
```

```python
model.evaluate(X_train,Y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y,Y_pread)
cm
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True)
```

```python
labels = ["True Neg","False Pos","False Neg","True Pos"]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
```

```python
group_names = ["True Neg","False Pos","False Neg","True Pos"]
group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)
sns.heatmap(cm, annot=labels, fmt='', cmap='Blues')
```

```python
sns.heatmap(cm/np.sum(cm), annot=True, 
            fmt='.2%', cmap='Blues')
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
print("Accuracy score: ",accuracy_score(Y,Y_pread).round(2))
print("Sensitivity score: ",recall_score(Y,Y_pread).round(2))
print("Precision score: ",precision_score(Y,Y_pread).round(2))
print("F1score score: ",f1_score(Y,Y_pread).round(2))
```

```python
TN = cm[0,0]
FP = cm[0,1]
TNR = TN / (TN + FP)
print("Specificity score: ",TNR.round(2))
```
